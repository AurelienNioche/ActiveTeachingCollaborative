```python
delay = 1
forgetting_rate = 0.7
```

```python
np.random.seed(1112)

mu = bp.nodes.GaussianARD(0, 10, name='mu')
sg = bp.nodes.Gamma(1e-6, 1e-6, name='sg')
n_total = 10000
n_sample = 10
y = bp.nodes.GaussianARD(mu, sg, plates=(n_sample,),plates_multiplier=(n_total/n_sample,), name='y')

Q = bp.inference.VB(mu, sg, y)
Q.ignore_bound_checks = True
for i in range(n_total):
    data = np.random.normal(loc=5, scale=10, size=n_sample)
    Q['y'].observe(data)
    _ = Q.update('mu', 'sg')
    step = (i + delay) ** (-forgetting_rate)
    _ = Q.gradient_step('mu', 'sg', scale=step)

bp.plot.pdf(mu, np.linspace(-10, 20, 100))
plt.show()
```
